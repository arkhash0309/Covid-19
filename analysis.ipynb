## Covid-19 Analytics using Matplotlib library in Python

Let us import the necessary libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Let us read the dataset.

```python
df = pd.read_csv('dataset/case_time_series.csv')
df
```

This is a data frame with 105 rows and 7 columns.

Let us store the following in arrays.
- Date
- Daily number of confirmed cases
- Daily number of recovered cases
- Daily number of deceased cases

```python
date = df.iloc[61:,0].values
daily_confirmed = df.iloc[61:,1].values
daily_recovered = df.iloc[61:,3].values
daily_deceased = df.iloc[61:,5].values
```

```python
date
```

```python
plt.figure(figsize=(25, 8))  
plt.plot(date, daily_confirmed, marker='o', linestyle='-', color='hotpink')  
plt.xlabel("Date", color='black')  
plt.ylabel("Daily number of confirmed cases", color='black')  
plt.title("Daily Confirmed Cases against Date", color='black')  
plt.grid(True, color='black', linestyle='--', alpha=0.5)  
plt.tick_params(axis='x', colors='black')
plt.tick_params(axis='y', colors='black')
plt.show()
```

Now let us try to plot a Pie Chart.

```python
pie_slices = [62, 142, 195]
activities = ['Travel', 'Place Visit', 'Unknown']
colours = ['#4C8BE2','#00e061','#fe073a']
exp = [0.2,0.02,0.02]

plt.pie(pie_slices, labels=activities, textprops=dict(size='10', color='black'), radius=1, colors=colours, autopct='%2.2f%%', explode=exp, shadow=True, startangle=90)
plt.title('Transmission', color='black', fontsize=30)
```

Finally let us come up with a Bar Plot.

```python
df = pd.read_csv('dataset/district.csv')
df
```

```python
recovered =df.iloc[:30,5].values
deceased =df.iloc[:30,4].values
confirmed =df.iloc[:30,3].values
district_name =list(df.iloc[:30,0])
```

```python
plt.figure(figsize=(12,10))
ax=plt.axes()
 
ax.set_facecolor('black')
ax.grid(linewidth=0.4, color='#8f8f8f')
 
 
plt.xticks(rotation='vertical',
           size='2',
           color='white')#ticks of X
 
plt.yticks(size='20',color='white')
 
 
ax.set_xlabel('District',size=25,
              color='#4bb4f2')
ax.set_ylabel('No. of cases',size=25,
              color='#4bb4f2')
 
 
plt.tick_params(size=20,color='white')
 
 
ax.set_title('Maharashtra District wise breakdown\n',
             size=30,color='#28a9ff')
 
plt.bar(district_name,confirmed,label='re')
plt.bar(district_name,recovered,label='re',color='green')
plt.bar(district_name,deceased,label='re',color='red')
 
for i,j in zip(district_name,confirmed):
    ax.annotate(str(int(j)),
                xy=(i,j+3),
                color='white',
                size='15')
 
plt.legend(['Confirmed','Recovered','Deceased'],
           fontsize=20)
```
